```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('hotel_booking.csv')

print('Dataset Shape:', df.shape)
print('Column Names:', df.columns.tolist())
print('\nMissing Values per Column:\n', df.isnull().sum())
```

## Data Cleaning - Handling Missing Values

```python
missing_values = df.isnull().sum() / len(df)
high_missing_cols = missing_values[missing_values > 0.4].index.tolist()
df.drop(columns=high_missing_cols, inplace=True)
df.dropna(inplace=True)
print('Remaining Missing Values:', df.isnull().sum().sum())
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
print('Numerical Columns:', numerical_cols)
print('Categorical Columns:', categorical_cols)
```

# 2. Univariate Analysis - Numerical

```python
for col in numerical_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col} (Histogram + KDE)')
    plt.show()
```

```python
for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12,4))
    sns.violinplot(y=df[col], ax=axes[0])
    axes[0].set_title(f'Violin Plot of {col}')
    sns.boxplot(y=df[col], ax=axes[1])
    axes[1].set_title(f'Box Plot of {col}')
    plt.show()
```

# 2. Univariate Analysis - Categorical

```python
for col in categorical_cols:
    if df[col].nunique() <= 50:
        plt.figure(figsize=(6,3))
        sns.countplot(x=df[col], order=df[col].value_counts().index)
        plt.xticks(rotation=45)
        plt.title(f'Count Plot of {col}')
        plt.show()
```

```python
for col in categorical_cols:
    if df[col].nunique() <= 50:  # Only for categories with fewer unique values
        plt.figure(figsize=(6,6))
        df[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm')
        plt.title(f'Pie Chart of {col}')
        plt.ylabel('')
        plt.show()
```

# 3. Bivariate Analysis

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numerical Columns)')
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Selecting numerical columns with reasonable cardinality
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
valid_numerical_cols = [col for col in numerical_cols if df[col].nunique() < 50]  # Exclude high-cardinality ones

# Generate pair plot
sns.pairplot(df[valid_numerical_cols])
plt.show()
# too many graphs must plot between desired according to need
```

```python

from scipy.stats import ttest_ind

# Selecting a categorical column with two groups for t-test
category_col = 'hotel'  # Example: Comparing city hotel vs resort hotel
num_col = 'adr'  # Example: Average daily rate comparison

group1 = df[df[category_col] == df[category_col].unique()[0]][num_col].dropna()
group2 = df[df[category_col] == df[category_col].unique()[1]][num_col].dropna()

# Perform t-test
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
print(f"T-Test Statistic: {t_stat}, P-value: {p_value}")
```

```python

from scipy.stats import f_oneway

# Selecting a categorical column with multiple groups for ANOVA (F-test)
category_col = 'market_segment'  # Example: Comparing ADR across different segments
num_col = 'adr'

# Extract data groups
groups = [df[df[category_col] == category][num_col].dropna() for category in df[category_col].unique() if len(df[df[category_col] == category]) > 10]

# Perform ANOVA F-test
f_stat, p_value = f_oneway(*groups)
print(f"F-Test Statistic: {f_stat}, P-value: {p_value}")
```
